С целью анализа причины пропусков посмотрим для каких платформ утеряны года

```python
# визуализируем долю пропусков от всех записей по каждой платформе
all_grouped_platform = df.groupby('platform')['name'].count().reset_index() # группировка по платформе всех записей
# группировка по платформам записей с пропуском в годе
nan_grouped_platform = df[df['year_of_release'].isna()].groupby('platform')['name'].count().reset_index() 
all_grouped_platform.merge(nan_grouped_platform, on='platform', how='left').sort_values(by='name_x').\
                   plot(kind='barh', x='platform', figsize=(17,8))  # объединение таблиц + сортировка + график
plt.legend(['Все записи','Записи с пропущенным годом'])
plt.title('Доля пропусков в годе выхода, группировка по платформам')
plt.show()
```

Видим, что пропуски года появились для игр с разных платформ. В основном их доля пропорциональна общему количеству игр на платформе: чем больше игр, тем больше пропусков. Есть одна платформа `2600`, которая имеет относительно большую долю пропусков, так как количество игр небольшое.
Мы не видим закономерности появления пропусков. Так как параметр важен для прогнозирования и общая доля пропусков всего лишь 1,6%, предлагаем удалить 269 записей.

```python
# удаляем строки с пропуском в годе
df = df.dropna(subset=['year_of_release'])
# проверим количество оставшихся строк
len(df)
```

#### Пропуски в `critic_score`

```python
# посчитаем, сколько строк, в которых пропуски встречаются сразу в тдвух столбцах
df[df['user_score'].isna() & df['rating'].isna()].shape[0]
```

Основная часть отсутствующих данных с оценками и рейтингом в одних и тех же записях.

Предположим, что эти данные отсутствуют для игр более ранних годов выхода, когда рейтинги были еще не введены, а оценки не распространены. Посмотрим, как распределяются пропуски по разным годам.

Известно, что компания ESRB, присвающая возрастную категорию (`rating`), была основана только в 1994 году, а данные в датасете начинаются с 1980

```python
# сгруппируем по годам, выведем количество записей в 4 столбцах, name возьмем для сравнения в качестве количества всех строк в годе
df.groupby('year_of_release').count().\
        plot(kind='bar', y=['name','critic_score', 'user_score', 'rating'], figsize=(17,5),\
            title='Распределение пропусков в столбцах `critic_score`, `user_score`, `rating` по годам',\
            )
plt.legend(['Все записи', 'critic_score', 'user_score', 'rating'])
plt.show()
```

\*разница с синим столбцом показывает долю пропусков.*

- в период 2007-2011 - пик по количеству вышедших игр, пропусков в оценке критиков становится заметно меньше, чем оценок пользователей и рейтинга.

Для платформы DS видим выбивающееся значение - 1985 год.

Всего одна запись c очевидно неправильным годом. Считаем ошибкой, удалим из датасета

```python
# удаление строки
df = df[df['name']!='Strongest Tokyo University Shogi DS']
# проверим
len(df)
```

```python
# посчитаем сколько лет каждая платформа имела какие-либо продажи (больше 0) и сколько больше 100 млн копий
list1=[]
dic1 = {}
for platform in list_of_large_platforms: # 
    dic1[platform] = []
    grouped = df[df['platform'] == platform].groupby('year_of_release')['all_sales'].sum()
    for item in grouped:
        if item > 0: # if item > 100:  - второй вариант на пике продаж
            dic1[platform].append(item) 
    list1.append(len(dic1[platform]))
print(list1)
f'Средняя продолжительность для 6 платформ {mean(list1)} лет'
```

В среднем продолжительность наиболее популярных платформ составляла:

- У самого распространенного жанра игр `Action` оценки средние, а у самых продаваемых `Sport` и `Shooter` относительно низкие.

Оценки пользователей выше у платформ и жанров, которые более популярны на японском рынке.



### Проверка гипотезы: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**Н_0:** *Средний пользовательский рейтинг ('user_score') платформы `XOne` = cредний пользовательский рейтинг ('user_score') платформы `PC `*  
**H_a:** *Средний пользовательский рейтинг ('user_score') платформы `XOne` ≠ cредний пользовательский рейтинг ('user_score') платформы `PC `*


```python
# создадим срезы из актуального периода игр с платформ XOne и PC, исключим пропуски для корректной работы метода
xone_user_score = actual_df[(actual_df['platform']=='XOne')&(actual_df['user_score'].isna()==False)]['user_score']
pc_user_score = actual_df[(actual_df['platform']=='PC')&(actual_df['user_score'].isna()==False)]['user_score']
```

```python
# посмотрим визуально на распределения оценок по этим двум платформам
xone_user_score.hist(alpha=0.8)
pc_user_score.hist(alpha=0.7)
plt.title('Распределение оценок пользователей для платформ "XOne" и "PC"')
plt.legend(['XOne', 'PC'])
plt.show()
```

```python
# зададим уровень значимости 5%
alpha = 0.05

# вызов метода для проверки гипотезы
results = st.ttest_ind(xone_user_score, pc_user_score, equal_var=False)

# вывод значения p-value на экран 
print(f'pvalue {results.pvalue}')

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** При уровне значимости 5% мы не имеем оснований отвергнуть гипотезу о том, что средние пользовательские рейтинги платформ `XOne` и `PC` одинаковые при уровне значимости 5%

### Проверка гипотезы: Средние пользовательские рейтинги жанров `Action` и `Sports` разные.  

В этом случае по требованию метода за нулевую гипотезу мы должны взять равенство средних, а за альтернативную - предлогаемую гипотезу о разности.

**Н_0:** *Средний пользовательский рейтинг ('user_score') жанра `Action` = cредний пользовательский рейтинг ('user_score') жанра `Sports`*  
**H_a:** *Средний пользовательский рейтинг ('user_score') жанра `Action` ≠ cредний пользовательский рейтинг ('user_score') жанра `Sports`*

```python
# создадим срезы из актуального периода игр с платформ XOne и PC, исключим пропуски параметром nan_policy метода ттест
action_user_score = actual_df[(actual_df['genre']=='Action')]['user_score']
sports_user_score = actual_df[(actual_df['genre']=='Sports')]['user_score']
```

```python
# посмотрим визуально на распределения оценок по этим двум жанрам
action_user_score.hist(alpha=0.8)
sports_user_score.hist(alpha=0.7)
plt.title('Распределение оценок пользователей для жанроов "Action" и "Sports"')
plt.legend(['Action', 'Sports'])
plt.show()
```

```python
# зададим уровень значимости 5%
alpha = 0.05

# вызов метода для проверки гипотезы
results = st.ttest_ind(action_user_score, sports_user_score, equal_var=False, nan_policy='omit')

# вывод значения p-value на экран 
print(f'pvalue {results.pvalue}')

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** На имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров `Action` и `Sports` в пользу альтернативной гипотезы.
